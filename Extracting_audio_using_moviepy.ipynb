```python
import moviepy.editor
```

```python
video = moviepy.editor.VideoFileClip('Euro Exim Bank_ Facilitating Global Trade 2.mp4')
```

```python
audio = video.audio
```

```python
audio
```

```python
audio.write_audiofile('graham.wav')
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="AqeelShafy7/AudioSangraha-Audio_to_Text")
```

```python
text = pipe('graham_audio.wav')
```
