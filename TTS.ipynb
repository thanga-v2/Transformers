```python
import torch
from TTS.api import TTS
import gradio as gr

# Check if CUDA is available and set the device
device = "cuda" if torch.cuda.is_available() else "cpu"
print("CUDA Available:", torch.cuda.is_available())

def generate_audio(text="A journey of a thousand miles begins with a single step"):
    # Initialize TTS with a multilingual model
    tts = TTS(model_name='tts_models/multilingual/multi-dataset/xtts_v2')

    # Generate the audio file
    output_path = "outputs/result.wav"
    tts.tts_to_file(
        text=text,
        file_path=output_path,
        speaker="Ana Florence",  # Specify speaker
        language="en",          # Specify language
    )

    return output_path  # Return the path to the generated audio file

# Gradio Interface
demo = gr.Interface(
    fn=generate_audio,
    inputs=[gr.Text(label="Input Text")],
    outputs=[gr.Audio(label="Generated Audio")],
    title="Text-to-Speech Generator",
    description="Enter text and generate speech using the TTS library."
)

# Launch the Gradio application
demo.launch(debug=True)
```
