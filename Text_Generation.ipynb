```python
lines = [line.rstrip() for line in open('robert_frost.txt')]
lines = [line for line in open('robert_frost.txt') if len(line) > 0]
```

```python


# pipeline for text generation
```

```python
import torch
```

```python
torch.cuda.is_available()
```

```python
torch.cuda.current_device()
```

```python
text_gen = pipeline('text-generation', device = 0)
```

```python
# setting the seed will produce the same result on client machine as well.

set_seed(2444)
```

```python
lines[0]
```

```python
text_gen(lines[0])
```

```python
# to make it in a readable manner use pprint(_)

pprint(_)
```

```python
pprint(text_gen(lines[0], max_length=20))
```

```python
# increasing the num_return_sequqnces will give multiple samples.

pprint(text_gen(lines[0], num_return_sequences=3, max_length=30))
```

```python
def wrap(x):
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
out = text_gen(lines[0], max_length=30)
#print("Generated output \n", wrap(out))
out
```

```python
out[0]
```

```python
out[0]['generated_text']
```

```python
print(' Generated Output \n', wrap(out[0]['generated_text']))
```

```python
prev = 'Two roads diverged in a yellow wood, including one that the' + 'road leading to another intersection in the middle of the city'
```

```python
out = text_gen(prev + '\n' + lines[2], max_length=60)
out[0]['generated_text']
print(wrap(out[0]['generated_text']))
```
