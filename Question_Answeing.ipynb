```python
from transformers import pipeline
```

```python
qa = pipeline("question-answering")
```

```python
context = "Today I went to the store to purchase a carton of milk"
question = "What did I buy?"
```

```python
qa(context=context, question=question)
```

```python
context[38:54]
```

```python
qa(context=context, question="what is the capital of Germany?")
```

```python
context = "Out of all the countries, I prefer to live in Germany / Deutschland"
qa(context=context, question = "what is my favourite place to live?")
```

```python
context = "Out of all the countries, I prefer not to live in USA"
qa(context=context, question = "what is my favourite place to live in?")
```
